```python
def process_file(filename):
    with open(filename, 'r', encoding='utf-8') as file:
        data = json.load(file)

    dialogue_f = {"c1": data["c1"], "c2": data["c2"]}
    return dialogue_f

if __name__ == "__main__":
    filename = "/content/external-362-79685652429-20230417-143020-1681731020.4558.txt"
    dialogue_f = process_file(filename)

print(dialogue_f)
```

```python
def extract_text_only(dialogue):
    text_only = ""
    for dialog in dialogue.values():
        for segment in dialog:
            text_only += segment['text'] + '\n'
    return text_only.strip()


result_text = extract_text_only(dialogue_f)
print(result_text)
```
